```python
import pandas as pd 
titanic = pd.read_csv('train.csv')
print('Initial Input:')
titanic
```

```python
titanic['family_size'] = titanic['SibSp'] + titanic['Parch'] + 1
titanic['Age'].fillna(titanic['Age'].median(), inplace=True)
titanic['Age_Cat'] = pd.cut(titanic['Age'], bins=[0,12,20,40,120], labels=['Children','Teenage','Adult','Elder'])
columns_to_drop = ["Survived", "Pclass", "SibSp","Parch","Ticket","Fare","Cabin","Embarked"]
titanic.drop(columns_to_drop, axis=1, inplace=True)
print(' Final Output:')
titanic
```
